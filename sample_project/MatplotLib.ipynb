## Matplotlib example workflow`m

```python
#0. import matplotlib and get it ready for ploting jupyter
%matplotlib inline
import matplotlib.pyplot as plt
#1. Prepare data
x = [1,2,3,4]
y = [11,22,33,44]

#2. setup plot
fig, ax = plt.subplots(figsize=(10,10)) # height and width of plot

#3 plot the data
ax.plot(x, y)

#4. customise the plot
ax.set(title="Simple Plot",
       xlabel="x-axis",
       ylabel="y axis")

#4. save and share
fig.savefig("Numpy Images/sample-plot.png")
```

## OO method to replicate prev plot

```python
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target',
             ax=ax); #same as prev with bigger size

ax.set_xlim([45,100]); #sets the x axis view limits
```

```python
##OO method from scratch

fig, ax = plt.subplots(figsize=(10,6))

#plot data
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"]);

#customzie the plot
ax.set(title="Heart Disease and Cholestrol level",
       xlabel="Age",
       ylabel="Cholestrol");

#Add a legend
ax.legend(*scatter.legend_elements(), title="Target"); #this asks to unpack tha legend elements for the c elements ie Target from the scatter plot assigning each of them a unique color as below

#add a horizontal line representing mean for each target value
ax.axhline(over_50["chol"].mean(),
            linestyle="-."
          ) #it takes a cumpolsory parameter as the y axis point against which we draw the line 
#linestyle parameter defines the style of line we can also werite ls instead of linestyle, default : - (straight line without any gaps)
```

## Adding a subplot to above plot

```python
over_50.head()
```

```python
fig, (ax0, ax1)= plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10,10),
                              sharex = True #sharing the x axis among the plots in the fig
                             );

#title to the figure

fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");

#plotting data
scatter = ax0.scatter(x=over_50["age"],
                       y=over_50["chol"],
                       c=over_50["target"]);
scatter1 = ax1.scatter(x=over_50["age"],
                       y=over_50["thalach"],
                       c=over_50["target"]);

#customize ax0 and ax1
ax0.set(title ="Heart disease and cholestrol",
        ylabel="cholestrol");
ax1.set(title = "Heart Disease and Thalach",
        xlabel="Age",
        ylabel="Max Heart Rate");

#adding legend to axes
ax0.legend(*scatter.legend_elements(), title="target");
ax1.legend(*scatter1.legend_elements(), title="target");

#horizontal mean line to axes ax0 and ax1
ax0.axhline(over_50["chol"].mean(), ls="--");
ax1.axhline(over_50["thalach"].mean(), ls="-.");
```
